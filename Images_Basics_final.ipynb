```python
import cv2
#Read Image
img_cv2 = cv2.imread(path_to_codes+'movie-wall-e-ad.jpg')


print(type(img_cv2))
print(img_cv2.shape)
  
# Output Images
plt.imshow(img_cv2)
plt.show()


# OpenCV represents RGB images as multi-dimensional NumPy arraysbut in reverse order!

# This means that images are actually represented in BGR order rather than RGB!

# Theres an easy fix though.

# All we need to do is convert the image from BGR to RGB:

new_img_cv2 = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
plt.imshow(new_img_cv2)
plt.show()


# showing gray images

gray_img_cv2 = new_img_cv2[:,:,0]

# Output Images
plt.imshow(gray_img_cv2)
plt.show()


# Output Images
plt.imshow(gray_img_cv2,cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_cv2 = cv2.imread(path_to_codes+'movie-wall-e-ad.jpg')

row, col, ch = np.shape(img_cv2)
print(row, col, ch)
new_img1 = np.uint(np.zeros((row,col,ch)))

thr = 120

for i in range(0,row):
  for j in range(0,col):
    if img_cv2[i,j,0] > thr:
      new_img1[i,j,:] = 255
    else:
      new_img1[i,j,:] = 0


plt.imshow(img_cv2)
plt.show()

plt.imshow(new_img1)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_cv2 = cv2.imread(path_to_codes+'circles.jpg')

row, col, ch = np.shape(img_cv2)
print(row, col, ch)
new_img1 = np.uint(np.zeros((row,col,ch)))

thr = 80

for i in range(0,row):
  for j in range(0,col):
    temp = (img_cv2[i,j,0]+ img_cv2[i,j,1] +img_cv2[i,j,2])//3
    if temp > thr:
      new_img1[i,j,:] = 255
    else:
      new_img1[i,j,:] = 0


plt.imshow(img_cv2)
plt.show()

plt.imshow(new_img1)
plt.show()
```
